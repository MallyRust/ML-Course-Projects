```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#read file
df = pd.read_csv('Salaries.csv')
```

```python
df.info()
```

 **1. Calculating standard deviation and variance without using methods std() and *variance()**


```python
#task 1

data = ['BasePay', 'OvertimePay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'Year']

mean_value = df[data].mean()
deviation = df[data] - mean_value
squared_deviation = np.square(deviation)
sum_squares = squared_deviation.sum()
N = len(df[data])

variance = sum_squares/N
std_dev = np.sqrt(variance)

print("Standart Deviation:\n",std_dev)
print("\nVariance:\n",variance)
```

**2. Calculating standard deviation with std() and var() methods**

```python
std_dev2 = np.std(df[data])
var2 = np.var(df[data])

print("Standard Deviation:\n", std_dev2)
print("\nVariance:\n", var2)
```

**3. Calculating coeficient of covarience and range**

using formula:
CV = standard deviation / sample mean x 100

```python
coef_covar = (std_dev2/mean_value)*100
print("Coefficient of covariance:\n", coef_covar)

correlation_matrix = df[data].corr()

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()

min_value = df[data].min()
max_value = df[data].max()
range = max_value - min_value

print("\nRange:\n", range)
```

**4. Calculating the measures of central tendency**

```python
mean_value = df[data].mean()
median_value = df[data].median()
mode_value = df[data].mode()

print("Mean:\n", mean_value)
print("\nMedian:\n", median_value)
print("\nMode:\n", mode_value)
```

**5. Applying the describe method**

```python
print("Description:\n", df[data].describe())
```

**6. Building histograms**

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 12))

axes = axes.flatten()

for i, col in enumerate(data):
  axes[i].hist(df[col], edgecolor='black')
  axes[i].set_title(col)

plt.tight_layout()
```

**7.Building boxplots**

```python
red_circle = dict(markerfacecolor='red', marker='o', markeredgecolor='white')
fig, axs = plt.subplots(1, len(data), figsize=(20, 10))

for i, ax in enumerate(axs.flat):
    ax.boxplot(df[data[i]].dropna(), flierprops=red_circle)
    ax.set_title(data[i], fontsize=16, fontweight='bold')

plt.tight_layout()
```

**8. Calculating IQR (Interquirtile range)**

```python
sorted = df[data].sort_values(by=data, ascending=True)

q1 = sorted.quantile(0.25)
q3 = sorted.quantile(0.75)

iqr = q3 - q1

print("Interquartile range:\n",iqr)

```
